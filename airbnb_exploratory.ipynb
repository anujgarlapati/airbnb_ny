```python
#Univariate Analysis
plt.figure(figsize=(10,6))
sns.distplot(airbnb_df.price,color='g')
plt.title('Price Distribution',size=18)
plt.xlabel('Price',size=14)
plt.ylabel('Density',size=14)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.distplot(airbnb_df.number_of_reviews, color='r')
plt.title('Numbers of Reviews Distribution', size=18)
plt.xlabel('Reviews', size=14)
plt.ylabel('Count', size=14)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(airbnb_df.minimum_nights,color='y')
plt.title('Minimum Nights Distribution',size=18)
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.boxplot(airbnb_df.price)
plt.title('Price',size=14)
plt.show()
```

```python
Q1 = airbnb_df['price'].quantile(0.25)
Q3 = airbnb_df['price'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
airbnb_df[(airbnb_df['price']< Q1-1.5* IQR) | (airbnb_df['price']> Q3+1.5* IQR)]
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x = 'room_type', data = airbnb_df)
plt.title(' Number of Room Types',size=18)
plt.xlabel('Room Type',size=14)
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.scatterplot(x='room_type',y='price',color='r',data=airbnb_df)
plt.title('Price vs Room Types',size=18)
plt.xlabel('Room Types',size=14)
plt.ylabel('Price',size=14)
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.heatmap(airbnb_df.corr(),annot=True,square=True,
            cmap='RdBu',
            vmax=1,
            vmin=-1)
plt.title('Correlations Between Variables',size=18);
plt.xticks(size=15)
plt.yticks(size=15)
plt.show()
```
